```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import os
```

```python
#read the csv file using pandas
df=pd.read_csv('./AllSTPrecipitation.csv')
df.head()
```

```python
#changing the format of date using pandas .to_datetime
df['Time']=pd.to_datetime(df['Time'])
df.head()
```

```python
df1=df.set_index('Time')
df1.head()
```

```python
#resampling the monthly data and sum these all precipitation value to show the total monthly trend

monthly=df1.resample('M').sum()
monthly
```

```python
plot=monthly.plot(figsize=(25,10), marker='+')
plot.set_xlabel('Date')
plot.set_ylabel('Total Precipitation(in mm)')
plot.set_title('Precipitation Trend from 2015 to 2022 Among Different Stations')


#to save the plot in hd
plt.savefig('./images/sample.png',dpi=2000)
```

```python
#second way of plotting the graph using matplotlib
df2 = pd.DataFrame(df, columns=["Time", "Pokhara Airport", "Lumle"])
df2.plot(x='Time',y=['Pokhara Airport','Lumle'], kind='line')
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import os
```

```python
df = pd.read_csv("./HourlyDataset.csv")
df.shape
df[25130:]
```

```python
#Total Null data available in the Dataset
df.isnull().sum()
```

```python
#changing the string type of time to the date data type
df['Time']=pd.to_datetime(df['Time'])
#changing the format of the date
df['Time']= df["Time"].dt.strftime('%Y-%m-%d %H:%M:%S')
df[:10]
```

```python
data_range = pd.date_range(start="2019-11-11 06:00:00", end="2011-11-07 06:00:00", freq="1H")
df.reindex(data_range, fill_value= ' ')
```

```python
df
```

```python
#Importing the missing hours in the DataFrame
df['Time'] = pd.to_datetime(df['Time'],format='%Y-%m-%d %H:%M:%S')
df = df.set_index(['Time']).resample('1H').last().reset_index()
df
```
